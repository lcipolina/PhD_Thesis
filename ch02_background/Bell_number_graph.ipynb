```python
import matplotlib.pyplot as plt
import numpy as np

# Data
n_agents = np.arange(1, 11)
bell_numbers = [1, 2, 5, 15, 52, 203, 877, 4140, 21147, 115975]
linear = n_agents
exponential = 2 ** n_agents

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n_agents, bell_numbers, 'o-', label='Bell Number $B(n)$')
plt.plot(n_agents, linear, 's-', label='Linear $n$')
plt.plot(n_agents, exponential, '^-', label='Exponential $2^n$')

plt.yscale('log')
plt.xlabel('Number of Agents $n$')
plt.ylabel('Value (log scale)')
plt.title('Growth Comparison: Bell Numbers, Linear ($n$) & Exponential ($2^n$)')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.tight_layout()
plt.show()
```
